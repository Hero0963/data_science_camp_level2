```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
```

```python
display(df)
```

```python
# - 1. filtered by first column > 20?
filter_1 = df[0] > 20  
display(df[filter_1])
```

```python
# - 2. filtered by first column + second column > 50
filter_2 = (df[0] + df[1]) > 50  
display(df[filter_2])
```

```python
# - 3. filtered by first column < 30 or second column > 30
filter_3 = (df[0] < 30) | (df[1] > 30)  
display(df[filter_3])
```

```python
# - 4. filtered by total sum of row > 100
filter_4 = df.sum(axis=1) > 100  
display(df[filter_4])
```
