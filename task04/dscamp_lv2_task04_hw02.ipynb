#練習：請利用題目提供的資料，分別完成以下幾種策略的資料填補方法：

```python
import numpy as np
import pandas as pd
raw_data = {'first_name': ['Jason', np.nan, 'Tina', 'Jake', 'Amy'], 
        'last_name': ['Miller', np.nan, 'Ali', 'Milner', 'Cooze'], 
        'age': [42, np.nan, 36, 24, 73], 
        'sex': ['m', np.nan, 'f', 'm', 'f'], 
        'preTestScore': [4, np.nan, np.nan, 2, 3],
        'postTestScore': [25, np.nan, np.nan, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'sex', 'preTestScore', 'postTestScore'])
```

```python
display(df)
```

```python
df.isnull()
```

```python
# Drop missing observations
df_drop = df.dropna()
display(df_drop)
```

```python
# Drop rows where all cells in that row is NA
df_drop_all = df.dropna(how='all')
display(df_drop_all)
```

```python
# Create a new column full of missing values
df_copy = df.copy()
new_cloumn = np.nan
df_copy['new_cloumn'] = new_cloumn
display(df_copy)
```

```python
# Fill in missing data with zeros
df_fill_zero = df.fillna(0)
display(df_fill_zero)
```

```python
# Fill in missing in preTestScore with the mean value of preTestScore
df_fill_preTestScore_with_mean = df.fillna({'preTestScore': df['preTestScore'].mean()})
display(df_fill_preTestScore_with_mean)
```

```python
# Fill in missing in postTestScore with each sex’s mean value of postTestScore
# ref = https://blog.csdn.net/fengdu78/article/details/120008871
# ref = https://zhuanlan.zhihu.com/p/86350553
df_fill_postTestScore_with_each_sex_mean = df.copy()
df_fill_postTestScore_with_each_sex_mean['postTestScore'] = df.groupby('sex')['postTestScore'].transform(lambda x: x.fillna(x.mean()))
display(df_fill_postTestScore_with_each_sex_mean)
```
