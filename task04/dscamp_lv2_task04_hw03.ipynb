#練習：How to check if a dataframe has any missing values？Check if df has any missing values.

```python
import numpy as np
import pandas as pd

url = 'https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv'
df = pd.read_csv(url)
print(type(df))
display(df.head())
```

```python
# Count the number of missing values in each column and find the maximum number of missing values.
# ref = https://stackoverflow.com/questions/52086574/pandas-isna-and-isnull-what-is-the-difference  isna() == isnull()
count_missing_values = df.isna().sum()
print(type(count_missing_values))
display(count_missing_values)
```

```python
df_count_missing_values = pd.DataFrame([count_missing_values])
print(type(df_count_missing_values))
display(df_count_missing_values)

```

```python
df_count_missing_values = df_count_missing_values.rename(index={0: 'count_missing_values'})
display(df_count_missing_values)
```

```python
df_count_missing_values = df_count_missing_values.T
display(df_count_missing_values)
```

```python
print(df_count_missing_values["count_missing_values"].max())
idx = df_count_missing_values["count_missing_values"].idxmax()
print(idx)
```

```python
# How to replace missing values of `multiple numeric` columns with the mean?
df_fill_numeric_mean = df.copy()
df_fill_numeric_mean.fillna(df.mean(numeric_only=True), inplace=True)
display(df_fill_numeric_mean.head())
```

```python
# Calculate the average price of different manufacturers.
df_avg_price_by_manufacturer = df.groupby('Manufacturer')['Price'].mean()
display(df_avg_price_by_manufacturer)
```

```python
# How to replace missing values of `price` columns with the mean depending on its manufacturers?
df_fill_price_with_manufacturer_mean = df.copy()
df_fill_price_with_manufacturer_mean['Price'] = df.groupby('Manufacturer')['Price'].transform(lambda x: x.fillna(x.mean()))
display(df_fill_price_with_manufacturer_mean.head())
```
