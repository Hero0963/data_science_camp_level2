#練習：請根據提供的資料，將原始欄位的資料切分成由 n 個類別所組成欄位。

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.random(100)*100)

n = int(input("How many categories do you want to divide it into ? "))

df['binned'] = pd.qcut(df[0], q=n, labels=False)
display(df)
```

```python
grouped_data = df.groupby('binned')[0].describe()
display(grouped_data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

x = df[0]
y = df['binned']

plt.scatter(x, y)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of X and Y')
plt.show()
```
