# **1. 請問「資料清理」、「資料視覺化」與「特徵工程」在資料前處理中扮演的角色是什麼？**      
  a. 資料清理： 將原始資料進行處理，去除缺失值或錯誤值，並轉成適合模型學習的資料型態。  

  b. 資料視覺化: 將資料以圖表的方式呈現，幫助我們進一步了解資料，有助於對資料進行初步的分析與探索，例如了解資料分佈以及變數之間的相關性。  

  c. 特徵工程: 對原始資料進行轉換與組合，從中提取出適合模型學習的特徵，目的是讓模型表現更佳、降低計算量、提供可解釋性等。
