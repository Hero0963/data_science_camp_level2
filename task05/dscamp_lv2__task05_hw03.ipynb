#練習：請根據提供的 `iris` 資料，完成以下操作：
* 1. 利用資料視覺化觀察 sepal 和 petal 四個欄位的關係。 
* 2. 利用相關係數觀察 sepal 和 petal 四個欄位的關係。  

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()

df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
display(df.head(200))
```

```python
print(iris.DESCR) 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
 
 
plt.scatter(df['sepal length (cm)'] , df['sepal width (cm)' ])
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('sepal length vs sepal width')
plt.show()
```

```python
plt.scatter(df['petal length (cm)'] , df['petal width (cm)' ])
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.title('petal length vs petal width')
plt.show()
```

```python
# ref = https://blog.csdn.net/hurry0808/article/details/78573585

pd.plotting.scatter_matrix(df , alpha = 0.7,  figsize=(14 , 8) , diagonal = "kde")
plt.show()
```

```python
# ref = https://reurl.cc/n7G31n

corr_matrix = df.corr()

plt.figure(figsize=(8, 8))
sns.heatmap(corr_matrix , annot = True ,fmt = '.2f')
plt.title('correlation matrix of sepal and petal features')
plt.show()
```
