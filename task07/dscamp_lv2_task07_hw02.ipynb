#練習：請將給定的原始資料計算出「每一個 Team」對應「每一個 Year」出現的次數：

```python
import numpy as np
import pandas as pd

ipl_data = {
    'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
    'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
    'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
    'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
    'Points':[876,789,863,673,741,812,756,788,694,701,804,690]
}
df = pd.DataFrame(ipl_data)
display(df)
```

```python
# ref = https://zhuanlan.zhihu.com/p/31952948
df_team_year_count = df.pivot_table(index = 'Year' , columns = 'Team' , values = 'Rank' , aggfunc = 'count')
# print(type(df_team_year_count))
display(df_team_year_count)
```
